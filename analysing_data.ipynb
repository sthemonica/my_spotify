```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go

from plotly.express import box

%matplotlib inline
```

```python
dados = pd.read_csv('spotify.csv')
dados.head()
```

```python
dados = dados.drop('Unnamed: 0', axis=1)
dados.head()
```

```python
dados = dados.drop('danceability.1', axis=1)
dados.head()
```

```python
import seaborn as sns

#ax = sns.violinplot(x=dados['danceability'], data=dados)

f, axes = plt.subplots(7, 1)

ax = sns.violinplot( data=dados ,x="danceability", linewidth = 0.6, inner = 'point', scale= 'width', ax=axes[0])
ax = sns.violinplot( data=dados ,x="acousticness", linewidth = 0.6, inner = 'point', scale= 'width', ax=axes[1])
ax = sns.violinplot( data=dados ,x="energy", linewidth = 0.6, inner = 'point', scale= 'width', ax=axes[2])
ax = sns.violinplot( data=dados ,x="instrumentalness", linewidth = 0.6, inner = 'point', scale= 'width', ax=axes[3])
ax = sns.violinplot( data=dados ,x="liveness", linewidth = 0.6, inner = 'point', scale= 'width', ax=axes[4])
ax = sns.violinplot( data=dados ,x="loudness", linewidth = 0.6, inner = 'point', scale= 'width', ax=axes[5])
ax = sns.violinplot( data=dados ,x="speechiness", linewidth = 0.6, inner = 'point', scale= 'width', ax=axes[6])
#ax = sns.violinplot( data=dados ,x="length", linewidth = 0.6, inner = 'point', scale= 'width', ax=axes[7])
```
